```python
df = pd.read_csv("Data.csv")
```

```python
df.head()
```

```python
top_rank_team = ['Australia','Sri Lanka','India', 'Bangladesh', 'New Zealand', 'South Africa', 'England',
       'West Indies', 'Afghanistan', 'Pakistan','Zimbabwe']
```

```python
df = df[df['batting_team'].isin(top_rank_team)]
```

```python
df = df[df['bowling_team'].isin(top_rank_team)]
```

```python
df.columns
```

```python
df['total_runs'] = df.apply(lambda x : (x.runs_off_bat + x.extras),axis=1)
```

```python
df.head()
```

```python
df=df.sort_values(['match_id', 'innings'], ascending=[True, True])
```

```python
df.head()
```

```python
df['overs'] = df.ball.apply(lambda x : x)
```

```python
df['over'] = df['ball'].astype(str)
```

```python
df['over'] = df.over.str[0:-1].str.replace('.' ,'')
```

```python
df['ball'] = df['ball'].astype(str).str[-1:]
```

```python
df=df[['match_id','innings','batting_team','bowling_team','overs','over','ball','runs_off_bat','total_runs','player_dismissed']]
```

```python
df['total'] = df.groupby(['match_id', 'innings'])['total_runs'].transform(sum)
```

```python
df['cumulative_total'] = df.groupby(['match_id', 'innings'])['total_runs'].apply(lambda x : x.cumsum())
```

```python
df['is_out'] = df.player_dismissed.apply(lambda x : 1 if type(x) == type('str') else 0)
df['player_dismissed'] = df.player_dismissed.apply(lambda x : 1 if type(x) == type('str') else 0)
df['total_wickets'] = df.groupby(['match_id', 'innings'])['is_out'].transform(sum)
df['cumulative_wicket'] = df.groupby(['match_id', 'innings'])['is_out'].apply(lambda x : x.cumsum())
```

```python
temp = df.groupby(['match_id', 'innings'])['total_runs'].rolling(min_periods = 1 , window = 30).sum().reset_index()
```

```python
df['prev_30Balls_runs'] = temp['total_runs'].to_list()
```

```python
tmp = df.groupby(['match_id', 'innings'])['is_out'].rolling(min_periods = 1 , window = 30).sum().reset_index()
```

```python
df['prev_30Balls_wickets'] = tmp['is_out'].to_list()
```

```python
df['prev_30Balls_Dots'] = df.total_runs.apply(lambda x : 1 if x == 0 else 0)
```

```python
tmp1 = df.groupby(['match_id', 'innings'])['prev_30Balls_Dots'].rolling(min_periods = 1 , window = 30).sum().reset_index()
```

```python
df['prev_30Balls_Dots'] = tmp1['prev_30Balls_Dots'].to_list()
```

```python
df['prev_30Balls_boundaries'] = df.runs_off_bat.apply(lambda x : 1 if x >= 4 else 0)
tmp1 = df.groupby(['match_id', 'innings'])['prev_30Balls_boundaries'].rolling(min_periods = 1 , window = 30).sum().reset_index()
df['prev_30Balls_boundaries'] = tmp1['prev_30Balls_boundaries'].to_list()
```

```python
convert_dict = {'prev_30Balls_runs' : int ,
               'prev_30Balls_wickets' : int,
               'prev_30Balls_Dots' : int,
               'prev_30Balls_boundaries' : int}
```

```python
df=df.astype(convert_dict)
df.head()
```

```python
df.to_csv("feature_data.csv" , index = None)
```
