```python
def find_phase(overs):
    if overs < 6:
        return 'Powerplay'
    elif overs > 15:
        return 'Death Overs'
    else:
        return 'Middle Overs'
```

```python
main_df['Phase'] = main_df['over'].apply(lambda x : find_phase(x))
main_df['Dots'] = main_df.batsman_runs.apply(lambda x : 1 if x == 0 else 0)
main_df['Ones'] = main_df.batsman_runs.apply(lambda x : 1 if x == 1 else 0)
main_df['Twos'] = main_df.batsman_runs.apply(lambda x : 1 if x == 2 else 0)
main_df['Threes'] = main_df.batsman_runs.apply(lambda x : 1 if x == 3 else 0)
main_df['Fours'] = main_df.batsman_runs.apply(lambda x : 1 if x == 4 else 0)
main_df['Sixes'] = main_df.batsman_runs.apply(lambda x : 1 if x == 6 else 0)
main_df['Fives'] = main_df.batsman_runs.apply(lambda x : 1 if (x == 5)  else 0)
main_df['Others'] = main_df.batsman_runs.apply(lambda x : 1 if (x == 7)  else 0)
main_df['isOut'] = main_df.player_dismissed.apply(lambda x : 1 if type(x) == type('str') else 0)
main_df = main_df[(main_df.inning == 1) | (main_df.inning == 2)]

# pd.DataFrame(main_df.groupby('batsman')['season'])
# main_df.to_csv('IplDataModified.csv')
```

```python
def balls_per_dismissals(balls,dismissals):
    if(dismissals == 0):
        return balls/1
    else:
        return balls/dismissals


def balls_per_boundaries(balls,boundaries):
    if(boundaries == 0):
        return balls/1
    else:
        return balls/boundaries
    
def average(runs,dismissals):
    if(boundaries == 0):
        return runs/1
    else:
        return runs/dismissals

def batsman_statistics(main_df):
    main_df = main_df[main_df.isWide == 0]
    dismiss_df = pd.DataFrame(main_df.groupby('player_dismissed')['isOut'].sum().reset_index().rename(columns={'isOut' : 'Dismissal'}))
    dismiss_df['batsman'] = dismiss_df.player_dismissed.apply(lambda x:x)
    Batsman_runs =pd.DataFrame(main_df.groupby(['batsman', 'matchId'])['batsman_runs'].sum().reset_index()).groupby(['batsman'])['batsman_runs'].sum().reset_index().rename(columns={'batsman_runs':'Batsman_runs'})
    Balls_Faced = pd.DataFrame(main_df.groupby('batsman')['ball'].count()).reset_index().rename(columns={'ball' : 'Balls_Faced'})
#     Dismissals = pd.DataFrame(main_df.groupby('batsman')['isOut'].sum()).rename(columns={'isOut' : 'Dismissals'})
    Innings = pd.DataFrame(main_df.groupby('batsman')['matchId'].apply(lambda x : len((np.unique(x))))).reset_index().rename(columns={'matchId':'No_of_Innings'})
    Dots = pd.DataFrame(main_df.groupby('batsman')['Dots'].sum()).reset_index()
    Ones = pd.DataFrame(main_df.groupby('batsman')['Ones'].sum()).reset_index()
    Twos = pd.DataFrame(main_df.groupby('batsman')['Twos'].sum()).reset_index()
    Threes = pd.DataFrame(main_df.groupby('batsman')['Threes'].sum()).reset_index()
    Fours = pd.DataFrame(main_df.groupby('batsman')['Fours'].sum()).reset_index()
    Fives = pd.DataFrame(main_df.groupby('batsman')['Fives'].sum()).reset_index()
    Sixes = pd.DataFrame(main_df.groupby('batsman')['Sixes'].sum()).reset_index()
    Others = pd.DataFrame(main_df.groupby('batsman')['Others'].sum()).reset_index()
    Batsman_stats_df = pd.merge(Innings , Batsman_runs , on = 'batsman').merge(Balls_Faced , on = 'batsman').merge(Dots , on = 'batsman').merge(Ones , on = 'batsman').merge(Twos , on = 'batsman').merge(Threes , on = 'batsman').merge(Fours , on = 'batsman').merge(Fives , on = 'batsman').merge(Sixes , on = 'batsman').merge(Others , on = 'batsman')
    Batsman_stats_df= pd.merge(Batsman_stats_df,dismiss_df[['batsman','Dismissal']],on='batsman', how='left')

    Batsman_stats_df['StrikeRate'] = Batsman_stats_df.apply(lambda x : 100* (x['Batsman_runs'] / x['Balls_Faced']),  axis = 1)
    Batsman_stats_df['Average'] = Batsman_stats_df.apply(lambda x : x['Batsman_runs']/x['Dismissal'],  axis = 1)
    Batsman_stats_df['RunsPerInnings'] = Batsman_stats_df.apply(lambda x : (x['Batsman_runs'] / x['No_of_Innings']),  axis = 1)
    Batsman_stats_df['BallsPerDismissal'] = Batsman_stats_df.apply(lambda x : balls_per_dismissals(x['Balls_Faced'] ,x['Dismissal']),  axis = 1)
    Batsman_stats_df['BallsPerBoundry'] = Batsman_stats_df.apply(lambda x : balls_per_boundaries(x['Balls_Faced'] ,x['Fours']+x['Sixes']),  axis = 1)
    Batsman_stats_df['Dot_percentage'] = Batsman_stats_df.apply(lambda x: x['Dots']/x['Balls_Faced'], axis = 1)
    return Batsman_stats_df

def req_batsman_statistics(main_df  ,stadium,phase,opposition ):
    main_df = main_df[main_df.isWide == 0]
    
    main_df = main_df[main_df['venue'] == stadium]   
    main_df = main_df[main_df['Phase'] == phase]  
    main_df = main_df[main_df['bowling_team'] == opposition]  
    main_df.reset_index(inplace = True, drop = True)
    
    dismiss_df = pd.DataFrame(main_df.groupby('player_dismissed')['isOut'].sum().reset_index().rename(columns={'isOut' : 'Dismissal'}))
    dismiss_df['batsman'] = dismiss_df.player_dismissed.apply(lambda x:x)
    Batsman_runs =pd.DataFrame(main_df.groupby(['batsman', 'matchId'])['batsman_runs'].sum().reset_index()).groupby(['batsman'])['batsman_runs'].sum().reset_index().rename(columns={'batsman_runs':'Batsman_runs'})
    Balls_Faced = pd.DataFrame(main_df.groupby('batsman')['ball'].count()).reset_index().rename(columns={'ball' : 'Balls_Faced'})
#     Dismissals = pd.DataFrame(main_df.groupby('batsman')['isOut'].sum()).rename(columns={'isOut' : 'Dismissals'})
    Innings = pd.DataFrame(main_df.groupby('batsman')['matchId'].apply(lambda x : len((np.unique(x))))).reset_index().rename(columns={'matchId':'No_of_Innings'})
    Dots = pd.DataFrame(main_df.groupby('batsman')['Dots'].sum()).reset_index()
    Ones = pd.DataFrame(main_df.groupby('batsman')['Ones'].sum()).reset_index()
    Twos = pd.DataFrame(main_df.groupby('batsman')['Twos'].sum()).reset_index()
    Threes = pd.DataFrame(main_df.groupby('batsman')['Threes'].sum()).reset_index()
    Fours = pd.DataFrame(main_df.groupby('batsman')['Fours'].sum()).reset_index()
    Fives = pd.DataFrame(main_df.groupby('batsman')['Fives'].sum()).reset_index()
    Sixes = pd.DataFrame(main_df.groupby('batsman')['Sixes'].sum()).reset_index()
    Others = pd.DataFrame(main_df.groupby('batsman')['Others'].sum()).reset_index()
    Batsman_stats_df = pd.merge(Innings , Batsman_runs , on = 'batsman').merge(Balls_Faced , on = 'batsman').merge(Dots , on = 'batsman').merge(Ones , on = 'batsman').merge(Twos , on = 'batsman').merge(Threes , on = 'batsman').merge(Fours , on = 'batsman').merge(Fives , on = 'batsman').merge(Sixes , on = 'batsman').merge(Others , on = 'batsman')
    Batsman_stats_df= pd.merge(Batsman_stats_df,dismiss_df[['batsman','Dismissal']],on='batsman', how='left')

    Batsman_stats_df['StrikeRate'] = Batsman_stats_df.apply(lambda x : 100* (x['Batsman_runs'] / x['Balls_Faced']),  axis = 1)
    Batsman_stats_df['Average'] = Batsman_stats_df.apply(lambda x : x['Batsman_runs']/x['Dismissal'],  axis = 1)
    Batsman_stats_df['RunsPerInnings'] = Batsman_stats_df.apply(lambda x : (x['Batsman_runs'] / x['No_of_Innings']),  axis = 1)
    Batsman_stats_df['BallsPerDismissal'] = Batsman_stats_df.apply(lambda x : balls_per_dismissals(x['Balls_Faced'] ,x['Dismissal']),  axis = 1)
    Batsman_stats_df['BallsPerBoundry'] = Batsman_stats_df.apply(lambda x : balls_per_boundaries(x['Balls_Faced'] ,x['Fours']+x['Sixes']),  axis = 1)
    Batsman_stats_df['Dot_percentage'] = Batsman_stats_df.apply(lambda x: x['Dots']/x['Balls_Faced'], axis = 1)
    return Batsman_stats_df


def req_batsman_statistics_scout(main_df ,phase ):
    main_df = main_df[main_df.isWide == 0]
    
#     main_df = main_df[main_df['venue'] == stadium]   
    main_df = main_df[main_df['Phase'] == phase]  
#     main_df = main_df[main_df['bowling_team'] == opposition]  
    main_df.reset_index(inplace = True, drop = True)
    
    dismiss_df = pd.DataFrame(main_df.groupby('player_dismissed')['isOut'].sum().reset_index().rename(columns={'isOut' : 'Dismissal'}))
    dismiss_df['batsman'] = dismiss_df.player_dismissed.apply(lambda x:x)
    Batsman_runs =pd.DataFrame(main_df.groupby(['batsman', 'matchId'])['batsman_runs'].sum().reset_index()).groupby(['batsman'])['batsman_runs'].sum().reset_index().rename(columns={'batsman_runs':'Batsman_runs'})
    Balls_Faced = pd.DataFrame(main_df.groupby('batsman')['ball'].count()).reset_index().rename(columns={'ball' : 'Balls_Faced'})
#     Dismissals = pd.DataFrame(main_df.groupby('batsman')['isOut'].sum()).rename(columns={'isOut' : 'Dismissals'})
    Innings = pd.DataFrame(main_df.groupby('batsman')['matchId'].apply(lambda x : len((np.unique(x))))).reset_index().rename(columns={'matchId':'No_of_Innings'})
    Dots = pd.DataFrame(main_df.groupby('batsman')['Dots'].sum()).reset_index()
    Ones = pd.DataFrame(main_df.groupby('batsman')['Ones'].sum()).reset_index()
    Twos = pd.DataFrame(main_df.groupby('batsman')['Twos'].sum()).reset_index()
    Threes = pd.DataFrame(main_df.groupby('batsman')['Threes'].sum()).reset_index()
    Fours = pd.DataFrame(main_df.groupby('batsman')['Fours'].sum()).reset_index()
    Fives = pd.DataFrame(main_df.groupby('batsman')['Fives'].sum()).reset_index()
    Sixes = pd.DataFrame(main_df.groupby('batsman')['Sixes'].sum()).reset_index()
    Others = pd.DataFrame(main_df.groupby('batsman')['Others'].sum()).reset_index()
    Batsman_stats_df = pd.merge(Innings , Batsman_runs , on = 'batsman').merge(Balls_Faced , on = 'batsman').merge(Dots , on = 'batsman').merge(Ones , on = 'batsman').merge(Twos , on = 'batsman').merge(Threes , on = 'batsman').merge(Fours , on = 'batsman').merge(Fives , on = 'batsman').merge(Sixes , on = 'batsman').merge(Others , on = 'batsman')
    Batsman_stats_df= pd.merge(Batsman_stats_df,dismiss_df[['batsman','Dismissal']],on='batsman', how='left')

    Batsman_stats_df['StrikeRate'] = Batsman_stats_df.apply(lambda x : 100* (x['Batsman_runs'] / x['Balls_Faced']),  axis = 1)
    Batsman_stats_df['Average'] = Batsman_stats_df.apply(lambda x : x['Batsman_runs']/x['Dismissal'],  axis = 1)
    Batsman_stats_df['RunsPerInnings'] = Batsman_stats_df.apply(lambda x : (x['Batsman_runs'] / x['No_of_Innings']),  axis = 1)
    Batsman_stats_df['BallsPerDismissal'] = Batsman_stats_df.apply(lambda x : balls_per_dismissals(x['Balls_Faced'] ,x['Dismissal']),  axis = 1)
    Batsman_stats_df['BallsPerBoundry'] = Batsman_stats_df.apply(lambda x : balls_per_boundaries(x['Balls_Faced'] ,x['Fours']+x['Sixes']),  axis = 1)
    Batsman_stats_df['Dot_percentage'] = Batsman_stats_df.apply(lambda x: x['Dots']/x['Balls_Faced'], axis = 1)
    return Batsman_stats_df
```

# TOPSIS

## Attributes Weight

```python
Wt_StrikeRate, Wt_RunsPerInnings, Wt_BallsPerDismissal, Wt_Dot_percentage = 0.13,0.27,0.16,0.45
```

## Step 1 – standardize the decision matrix.

```python
Req_batsman_stats['calc_SR'] = Req_batsman_stats.StrikeRate.apply(lambda x : x*x)
Req_batsman_stats['calc_RPI'] = Req_batsman_stats.RunsPerInnings.apply(lambda x : x*x)
Req_batsman_stats['calc_BPD'] = Req_batsman_stats.BallsPerDismissal.apply(lambda x : x*x)
Req_batsman_stats['calc_DP'] = Req_batsman_stats.Dot_percentage.apply(lambda x : x*x)

Sqrt_calc_SR, Sqrt_calc_RPI,Sqrt_calc_BPD,Sqrt_calc_DP = np.sqrt(Req_batsman_stats[['calc_SR' ,'calc_RPI','calc_BPD','calc_DP' ]].sum(axis=0))

Req_batsman_stats['calc_SR'] = Req_batsman_stats['StrikeRate'].apply(lambda x : x/Sqrt_calc_SR)
Req_batsman_stats['calc_RPI'] = Req_batsman_stats['RunsPerInnings'].apply(lambda x : x/Sqrt_calc_RPI)
Req_batsman_stats['calc_BPD'] = Req_batsman_stats['BallsPerDismissal'].apply(lambda x : x/Sqrt_calc_BPD)
Req_batsman_stats['calc_DP'] = Req_batsman_stats['Dot_percentage'].apply(lambda x : x/Sqrt_calc_DP)
```

## Step 2 - Construct weighted standardized decision matrix by multiplying attributes weight                      to each 

```python
Req_batsman_stats['calc_SR'] = Req_batsman_stats['calc_SR'].apply(lambda x : x*Wt_StrikeRate)
Req_batsman_stats['calc_RPI'] = Req_batsman_stats['calc_RPI'].apply(lambda x : x*Wt_RunsPerInnings)
Req_batsman_stats['calc_BPD'] = Req_batsman_stats['calc_BPD'].apply(lambda x : x*Wt_BallsPerDismissal)
Req_batsman_stats['calc_DP'] = Req_batsman_stats['calc_DP'].apply(lambda x : x*Wt_Dot_percentage)
```

## Step 3 – Determine ideal solution and negative ideal solution, i.e., best and worst (max and min values)

```python
best_SR , worst_SR = np.max(Req_batsman_stats['calc_SR'] ),np.min(Req_batsman_stats['calc_SR'] ) 
best_RPI , worst_RPI = np.max(Req_batsman_stats['calc_RPI'] ),np.min(Req_batsman_stats['calc_RPI'] ) 
best_BPD , worst_BPD = np.max(Req_batsman_stats['calc_BPD'] ),np.min(Req_batsman_stats['calc_BPD'] ) 
best_DP , worst_DP = np.min(Req_batsman_stats['calc_DP'] ),np.max(Req_batsman_stats['calc_DP'] ) 
```

## Step 4 – Determine separation from ideal solution. Si*

```python
Req_batsman_stats['best_SR'] = Req_batsman_stats['calc_SR'].apply(lambda x : (x-best_SR)*(x-best_SR))
Req_batsman_stats['best_RPI'] = Req_batsman_stats['calc_RPI'].apply(lambda x : (x-best_RPI)*(x-best_RPI))
Req_batsman_stats['best_BPD'] = Req_batsman_stats['calc_BPD'].apply(lambda x : (x-best_BPD)*(x-best_BPD))
Req_batsman_stats['best_DP'] = Req_batsman_stats['calc_DP'].apply(lambda x : (x-best_DP)*(x-best_DP))

Req_batsman_stats['best_Sqrt'] = np.sqrt(Req_batsman_stats.apply(lambda x :x['best_SR']+x['best_RPI']+x['best_BPD']+x['best_DP'],axis=1))
```

## Step 5 – Determine separation from negative ideal solution. Si

```python
Req_batsman_stats['worst_SR'] = Req_batsman_stats['calc_SR'].apply(lambda x : (x-worst_SR)*(x-worst_SR))
Req_batsman_stats['worst_RPI'] = Req_batsman_stats['calc_RPI'].apply(lambda x : (x-worst_RPI)*(x-worst_RPI))
Req_batsman_stats['worst_BPD'] = Req_batsman_stats['calc_BPD'].apply(lambda x : (x-worst_BPD)*(x-worst_BPD))
Req_batsman_stats['worst_DP'] = Req_batsman_stats['calc_DP'].apply(lambda x : (x-worst_DP)*(x-worst_DP))
Req_batsman_stats['worst_Sqrt'] = np.sqrt(Req_batsman_stats.apply(lambda x :x['worst_SR']+x['worst_RPI']+x['worst_BPD']+x['worst_DP'],axis=1))
```

## Step 6 – Determine relative closeness to ideal solution.

```python
Req_batsman_stats['Score'] = Req_batsman_stats.apply(lambda x :x['worst_Sqrt']/(x['best_Sqrt'] + x['worst_Sqrt'] ), axis =1)
```

```python
Req_batsman_stats.sort_values(by='Score' , ascending=False).head(10)
```

# Normalization => (x - Min_Val / Max_Val - Min_Val )

```python
Min_SR = min(Req_batsman_stats.StrikeRate)
Max_SR = max(Req_batsman_stats.StrikeRate)
Req_batsman_stats['Nor_SR'] = Req_batsman_stats.StrikeRate.apply(lambda x : ((x - Min_SR) / (Max_SR - Min_SR)))

Min_RPI = min(Req_batsman_stats.RunsPerInnings)
Max_RPI = max(Req_batsman_stats.RunsPerInnings)
Req_batsman_stats['Nor_RPI'] = Req_batsman_stats.RunsPerInnings.apply(lambda x : (x - Min_RPI) / (Max_RPI - Min_RPI))

Min_BPD = min(Req_batsman_stats.BallsPerDismissal)
Max_BPD = max(Req_batsman_stats.BallsPerDismissal)
Req_batsman_stats['Nor_BPD'] = Req_batsman_stats.BallsPerDismissal.apply(lambda x : (x - Min_BPD) / (Max_BPD - Min_BPD)) 

Min_DP = max(Req_batsman_stats.Dot_percentage)
Max_DP = min(Req_batsman_stats.Dot_percentage)
Req_batsman_stats['Nor_DP'] = Req_batsman_stats.Dot_percentage.apply(lambda x : (x - Min_DP) / (Max_DP - Min_DP))

Req_batsman_stats['Min_Max_Score'] = Req_batsman_stats.apply(lambda x : (x.Nor_SR * Wt_StrikeRate) + (x.Nor_RPI * Wt_RunsPerInnings) + (x.Nor_BPD * Wt_BallsPerDismissal ) + (x.Nor_DP * Wt_Dot_percentage), axis = 1)
Req_batsman_stats.sort_values('Min_Max_Score' , ascending=False).head(5)
```

```python
def Topsis_method_batting(Req_batsman_stats):
    Wt_StrikeRate, Wt_RunsPerInnings, Wt_BallsPerDismissal, Wt_Balls_per_Boundary = 0.13,0.27,0.16,0.45
    Req_batsman_stats['calc_SR'] = Req_batsman_stats.StrikeRate.apply(lambda x : x*x)
    Req_batsman_stats['calc_RPI'] = Req_batsman_stats.RunsPerInnings.apply(lambda x : x*x)
    Req_batsman_stats['calc_BPD'] = Req_batsman_stats.BallsPerDismissal.apply(lambda x : x*x)
    Req_batsman_stats['calc_BPB'] = Req_batsman_stats.BallsPerBoundry.apply(lambda x : x*x)

    Sqrt_calc_SR, Sqrt_calc_RPI,Sqrt_calc_BPD,Sqrt_calc_BPB = np.sqrt(Req_batsman_stats[['calc_SR' ,'calc_RPI','calc_BPD','calc_BPB' ]].sum(axis=0))

    Req_batsman_stats['calc_SR'] = Req_batsman_stats['StrikeRate'].apply(lambda x : x/Sqrt_calc_SR)
    Req_batsman_stats['calc_RPI'] = Req_batsman_stats['RunsPerInnings'].apply(lambda x : x/Sqrt_calc_RPI)
    Req_batsman_stats['calc_BPD'] = Req_batsman_stats['BallsPerDismissal'].apply(lambda x : x/Sqrt_calc_BPD)
    Req_batsman_stats['calc_BPB'] = Req_batsman_stats['BallsPerBoundry'].apply(lambda x : x/Sqrt_calc_BPB)
    
    Req_batsman_stats['calc_SR'] = Req_batsman_stats['calc_SR'].apply(lambda x : x*Wt_StrikeRate)
    Req_batsman_stats['calc_RPI'] = Req_batsman_stats['calc_RPI'].apply(lambda x : x*Wt_RunsPerInnings)
    Req_batsman_stats['calc_BPD'] = Req_batsman_stats['calc_BPD'].apply(lambda x : x*Wt_BallsPerDismissal)
    Req_batsman_stats['calc_BPB'] = Req_batsman_stats['calc_BPB'].apply(lambda x : x*Wt_Balls_per_Boundary)

    best_SR , worst_SR = np.max(Req_batsman_stats['calc_SR'] ),np.min(Req_batsman_stats['calc_SR'] ) 
    best_RPI , worst_RPI = np.max(Req_batsman_stats['calc_RPI'] ),np.min(Req_batsman_stats['calc_RPI'] ) 
    best_BPD , worst_BPD = np.max(Req_batsman_stats['calc_BPD'] ),np.min(Req_batsman_stats['calc_BPD'] ) 
    best_BPB , worst_BPB = np.min(Req_batsman_stats['calc_BPB'] ),np.max(Req_batsman_stats['calc_BPB'] ) 
    
    Req_batsman_stats['best_SR'] = Req_batsman_stats['calc_SR'].apply(lambda x : (x-best_SR)*(x-best_SR))
    Req_batsman_stats['best_RPI'] = Req_batsman_stats['calc_RPI'].apply(lambda x : (x-best_RPI)*(x-best_RPI))
    Req_batsman_stats['best_BPD'] = Req_batsman_stats['calc_BPD'].apply(lambda x : (x-best_BPD)*(x-best_BPD))
    Req_batsman_stats['best_BPB'] = Req_batsman_stats['calc_BPB'].apply(lambda x : (x-best_BPB)*(x-best_BPB))

    Req_batsman_stats['best_Sqrt'] = np.sqrt(Req_batsman_stats.apply(lambda x :x['best_SR']+x['best_RPI']+x['best_BPD']+x['best_BPB'],axis=1))
    
    Req_batsman_stats['worst_SR'] = Req_batsman_stats['calc_SR'].apply(lambda x : (x-worst_SR)*(x-worst_SR))
    Req_batsman_stats['worst_RPI'] = Req_batsman_stats['calc_RPI'].apply(lambda x : (x-worst_RPI)*(x-worst_RPI))
    Req_batsman_stats['worst_BPD'] = Req_batsman_stats['calc_BPD'].apply(lambda x : (x-worst_BPD)*(x-worst_BPD))
    Req_batsman_stats['worst_BPB'] = Req_batsman_stats['calc_BPB'].apply(lambda x : (x-worst_BPB)*(x-worst_BPB))
    
    Req_batsman_stats['worst_Sqrt'] = np.sqrt(Req_batsman_stats.apply(lambda x :x['worst_SR']+x['worst_RPI']+x['worst_BPD']+x['worst_BPB'],axis=1))
    
    Req_batsman_stats['Score'] = Req_batsman_stats.apply(lambda x :x['worst_Sqrt']/(x['best_Sqrt'] + x['worst_Sqrt'] ), axis =1)
    Req_batsman_stats[['batsman', 'No_of_Innings','Batsman_runs', 'Balls_Faced', 'Average','Dismissal','Dots', 'BallsPerBoundry','StrikeRate','RunsPerInnings','BallsPerDismissal','Score']].sort_values(by='Score' , ascending=False).head(10)
    
    return Req_batsman_stats
```

```python
def get_four_count(balls,fours):
    if(fours == 0):
        return 0
    else:
        return balls/fours
    
def get_six_count(balls,sixes):
    if(sixes == 0):
        return 0
    else:
        return balls/sixes
    
def average(runs,wickets):
    if(wickets == 0):
        return 0
    else:
        return runs/wickets

def strikeRate(balls,wickets):
    if(wickets == 0):
        return 0
    else:
        return balls/wickets

def get_Dot_percent(balls,dots):
    if(balls > 0):
        return 100 *(dots/balls)
    else:
        return 0
    
def economy(runs,balls):
    if(balls > 0):
        return 6*(runs/balls)
    else:
        return 0

def bowler_statistics(main_df):
    Runs_conceded = pd.DataFrame(main_df.groupby('bowler')['total_runs'].sum()).reset_index().rename(columns={'total_runs' : 'Runs_conceded'})
    Wickets = pd.DataFrame(main_df.groupby('bowler')['isBowlerWicket'].sum()).reset_index().rename(columns={'isBowlerWicket' : 'Wickets'})
    Balls_bowled = pd.DataFrame(main_df.groupby('bowler')['matchId'].count()).reset_index().rename(columns={'matchId' : 'Balls_bowled'})
    Innings = pd.DataFrame(main_df.groupby('bowler')['matchId'].apply(lambda x: len(list(np.unique(x))))).reset_index().rename(columns={'matchId':'No. of Innings'})
    Dots = pd.DataFrame(main_df.groupby('bowler')['Dots'].sum()).reset_index()
    Ones = pd.DataFrame(main_df.groupby('bowler')['Ones'].sum()).reset_index()
    Twos = pd.DataFrame(main_df.groupby('bowler')['Twos'].sum()).reset_index()
    Threes = pd.DataFrame(main_df.groupby('bowler')['Threes'].sum()).reset_index()
    Fours = pd.DataFrame(main_df.groupby('bowler')['Fours'].sum()).reset_index()
    Sixes = pd.DataFrame(main_df.groupby('bowler')['Sixes'].sum()).reset_index()
    Extras = pd.DataFrame(main_df.groupby('bowler')['extras'].sum()).reset_index().rename(columns={'extras' : 'Extras'})
    
    Bowler_stats_df = pd.merge(Innings , Wickets, on='bowler').merge(Balls_bowled , on='bowler').merge(Runs_conceded , on='bowler').merge(Dots , on='bowler').merge(Ones , on='bowler').merge(Twos , on='bowler').merge(Threes , on='bowler').merge(Fours , on='bowler').merge(Sixes , on='bowler').merge(Extras , on='bowler')
    Bowler_stats_df['BallsPerFour'] = Bowler_stats_df.apply(lambda x : get_four_count(x['Balls_bowled'] ,x['Fours']),  axis = 1)
    Bowler_stats_df['BallsPerSixes'] = Bowler_stats_df.apply(lambda x : get_six_count(x.Balls_bowled , x.Sixes), axis = 1)
    Bowler_stats_df['Bowling_Avg'] = Bowler_stats_df.apply(lambda x : average(x.Runs_conceded , x.Wickets), axis = 1)
    Bowler_stats_df['Bowling_StrikeRate'] = Bowler_stats_df.apply(lambda x : strikeRate(x.Balls_bowled , x.Wickets), axis = 1)
    Bowler_stats_df['Dot_percentage'] = Bowler_stats_df.apply(lambda x : get_Dot_percent(x.Balls_bowled , x.Dots), axis = 1)
    Bowler_stats_df['Economy'] = Bowler_stats_df.apply(lambda x : economy(x.Runs_conceded , x.Balls_bowled), axis = 1)
    
    return Bowler_stats_df
    
    

def Req_bowler_stats(main_df,phase,stadium):
    main_df = main_df[main_df.Phase == phase]
    main_df = main_df[main_df.venue == stadium]
    
    Runs_conceded = pd.DataFrame(main_df.groupby('bowler')['total_runs'].sum()).reset_index().rename(columns={'total_runs' : 'Runs_conceded'})
    Wickets = pd.DataFrame(main_df.groupby('bowler')['isBowlerWicket'].sum()).reset_index().rename(columns={'isBowlerWicket' : 'Wickets'})
    Balls_bowled = pd.DataFrame(main_df.groupby('bowler')['matchId'].count()).reset_index().rename(columns={'matchId' : 'Balls_bowled'})
    Innings = pd.DataFrame(main_df.groupby('bowler')['matchId'].apply(lambda x: len(list(np.unique(x))))).reset_index().rename(columns={'matchId':'No. of Innings'})
    Dots = pd.DataFrame(main_df.groupby('bowler')['Dots'].sum()).reset_index()
    Ones = pd.DataFrame(main_df.groupby('bowler')['Ones'].sum()).reset_index()
    Twos = pd.DataFrame(main_df.groupby('bowler')['Twos'].sum()).reset_index()
    Threes = pd.DataFrame(main_df.groupby('bowler')['Threes'].sum()).reset_index()
    Fours = pd.DataFrame(main_df.groupby('bowler')['Fours'].sum()).reset_index()
    Sixes = pd.DataFrame(main_df.groupby('bowler')['Sixes'].sum()).reset_index()
    Extras = pd.DataFrame(main_df.groupby('bowler')['extras'].sum()).reset_index().rename(columns={'extras' : 'Extras'})
    
    Bowler_stats_df = pd.merge(Innings , Wickets, on='bowler').merge(Balls_bowled , on='bowler').merge(Runs_conceded , on='bowler').merge(Dots , on='bowler').merge(Ones , on='bowler').merge(Twos , on='bowler').merge(Threes , on='bowler').merge(Fours , on='bowler').merge(Sixes , on='bowler').merge(Extras , on='bowler')
    Bowler_stats_df['BallsPerFour'] = Bowler_stats_df.apply(lambda x : get_four_count(x['Balls_bowled'] ,x['Fours']),  axis = 1)
    Bowler_stats_df['BallsPerSixes'] = Bowler_stats_df.apply(lambda x : get_six_count(x.Balls_bowled , x.Sixes), axis = 1)
    Bowler_stats_df['Bowling_Avg'] = Bowler_stats_df.apply(lambda x : average(x.Runs_conceded , x.Wickets), axis = 1)
    Bowler_stats_df['Bowling_StrikeRate'] = Bowler_stats_df.apply(lambda x : strikeRate(x.Balls_bowled , x.Wickets), axis = 1)
    Bowler_stats_df['Dot_percentage'] = Bowler_stats_df.apply(lambda x : get_Dot_percent(x.Balls_bowled , x.Dots), axis = 1)
    Bowler_stats_df['Economy'] = Bowler_stats_df.apply(lambda x : economy(x.Runs_conceded , x.Balls_bowled), axis = 1)
    
    return Bowler_stats_df
```

# TOPSIS

## Step 1 – standardize the decision matrix.

```python
Req_bowler_stats_PP['Bowl_calc_DP'] = Req_bowler_stats_PP.Dot_percentage.apply(lambda x : x*x)
Req_bowler_stats_PP['Bowl_calc_SR'] = Req_bowler_stats_PP.Bowling_StrikeRate.apply(lambda x : x*x)
Req_bowler_stats_PP['Bowl_calc_Eco'] = Req_bowler_stats_PP.Economy.apply(lambda x : x*x)
Req_bowler_stats_PP['Bowl_calc_Avg'] = Req_bowler_stats_PP.Bowling_Avg.apply(lambda x : x*x)

Sqrt_calc_dp, Sqrt_calc_sr,Sqrt_calc_eco,Sqrt_calc_avg = np.sqrt(Req_bowler_stats_PP[['Bowl_calc_DP' ,'Bowl_calc_SR','Bowl_calc_Eco','Bowl_calc_Avg' ]].sum(axis=0))

Req_bowler_stats_PP['Bowl_calc_DP'] = Req_bowler_stats_PP.Dot_percentage.apply(lambda x : x/Sqrt_calc_dp)
Req_bowler_stats_PP['Bowl_calc_SR'] = Req_bowler_stats_PP.Bowling_StrikeRate.apply(lambda x : x/Sqrt_calc_sr)
Req_bowler_stats_PP['Bowl_calc_Eco'] = Req_bowler_stats_PP.Economy.apply(lambda x : x/Sqrt_calc_eco)
Req_bowler_stats_PP['Bowl_calc_Avg'] = Req_bowler_stats_PP.Bowling_Avg.apply(lambda x : x/Sqrt_calc_avg)
```

## Step 2 - Construct weighted standardized decision matrix by multiplying attributes weight to each

```python
Req_bowler_stats_PP['Bowl_calc_DP'] = Req_bowler_stats_PP['Bowl_calc_DP'].apply(lambda x : x*Wt_Bowler_DP)
Req_bowler_stats_PP['Bowl_calc_SR'] = Req_bowler_stats_PP['Bowl_calc_SR'].apply(lambda x : x*Wt_Bowler_SR)
Req_bowler_stats_PP['Bowl_calc_Eco'] = Req_bowler_stats_PP['Bowl_calc_Eco'].apply(lambda x : x*Wt_Bowler_Eco)
Req_bowler_stats_PP['Bowl_calc_Avg'] = Req_bowler_stats_PP['Bowl_calc_Avg'].apply(lambda x : x*Wt_Bowler_Avg)
```

## Step 3 – Determine ideal solution and negative ideal solution, i.e., best and worst (max and min values)

```python
best_DP , worst_DP = np.max(Req_bowler_stats_PP['Bowl_calc_DP'] ),np.min(Req_bowler_stats_PP['Bowl_calc_DP'] ) 
best_SR , worst_SR = np.max(Req_bowler_stats_PP['Bowl_calc_SR'] ),np.min(Req_bowler_stats_PP['Bowl_calc_SR'] ) 
best_Eco , worst_Eco = np.max(Req_bowler_stats_PP['Bowl_calc_Eco'] ),np.min(Req_bowler_stats_PP['Bowl_calc_Eco'] ) 
best_Avg , worst_Avg = np.max(Req_bowler_stats_PP['Bowl_calc_Avg'] ),np.min(Req_bowler_stats_PP['Bowl_calc_Avg'] ) 
```

## Step 4 – Determine separation from ideal solution. Si*

```python
Req_bowler_stats_PP['best_DP'] = Req_bowler_stats_PP['Bowl_calc_DP'].apply(lambda x : (x-best_DP)*(x-best_DP))
Req_bowler_stats_PP['best_SR'] = Req_bowler_stats_PP['Bowl_calc_SR'].apply(lambda x : (x-best_SR)*(x-best_SR))
Req_bowler_stats_PP['best_Eco'] = Req_bowler_stats_PP['Bowl_calc_Eco'].apply(lambda x : (x-best_Eco)*(x-best_Eco))
Req_bowler_stats_PP['best_Avg'] = Req_bowler_stats_PP['Bowl_calc_Avg'].apply(lambda x : (x-best_Avg)*(x-best_Avg))

Req_bowler_stats_PP['best_Sqrt'] = np.sqrt(Req_bowler_stats_PP.apply(lambda x :x['best_DP']+x['best_SR']+x['best_Eco']+x['best_Avg'],axis=1))
```

## Step 5 – Determine separation from negative ideal solution. Si


```python
Req_bowler_stats_PP['worst_DP'] = Req_bowler_stats_PP['Bowl_calc_DP'].apply(lambda x : (x-worst_DP)*(x-worst_DP))
Req_bowler_stats_PP['worst_SR'] = Req_bowler_stats_PP['Bowl_calc_SR'].apply(lambda x : (x-worst_SR)*(x-worst_SR))
Req_bowler_stats_PP['worst_Eco'] = Req_bowler_stats_PP['Bowl_calc_Eco'].apply(lambda x : (x-worst_Eco)*(x-worst_Eco))
Req_bowler_stats_PP['worst_Avg'] = Req_bowler_stats_PP['Bowl_calc_Avg'].apply(lambda x : (x-worst_Avg)*(x-worst_Avg))
Req_bowler_stats_PP['worst_Sqrt'] = np.sqrt(Req_bowler_stats_PP.apply(lambda x :x['worst_DP']+x['worst_SR']+x['worst_Eco']+x['worst_Avg'],axis=1))
```

## Step 6 – Determine relative closeness to ideal solution.


```python
Req_bowler_stats_PP['Score'] = Req_bowler_stats_PP.apply(lambda x :x['worst_Sqrt']/(x['best_Sqrt'] + x['worst_Sqrt'] ), axis =1)
```

```python
def interactive_viz(main_df , match_id):

    matchID_df = pd.DataFrame( main_df[main_df.matchId == match_id])
    
    matchID_df.reset_index(inplace=True,drop=True)

    matchID_df_inns_1 = pd.DataFrame( matchID_df[matchID_df.inning == 1])
    matchID_df_inns_2 = pd.DataFrame( matchID_df[matchID_df.inning == 2])

    matchID_df_inns_1.reset_index(inplace=True,drop=True)
    matchID_df_inns_2.reset_index(inplace=True,drop=True)

    matchID_df_inns_1['cum_runs_1'] = 0
    matchID_df_inns_2['cum_runs_2'] = 0

    runs_count_1 = 0
    for i in range(len(matchID_df_inns_1)):
        runs_count_1 += matchID_df_inns_1.total_runs[i]
        matchID_df_inns_1['cum_runs_1'][i] = runs_count_1

    runs_count_2 = 0
    for i in range(len(matchID_df_inns_2)):
        runs_count_2 += matchID_df_inns_2.total_runs[i]
        matchID_df_inns_2['cum_runs_2'][i] = runs_count_2
    

    matchID_df_inns_1['Ball_No'] = [i for i in range(1, len(matchID_df_inns_1)+1)]
    matchID_df_inns_2['Ball_No'] = [i for i in range(1, len(matchID_df_inns_2)+1)]

    wicket_index_1 = list(matchID_df_inns_1[matchID_df_inns_1.isOut == 1].index)
    wicket_index_2 = list(matchID_df_inns_2[matchID_df_inns_2.isOut == 1].index)

    wicket_runs_1 = list(matchID_df_inns_1[matchID_df_inns_1.index.isin(wicket_index_1)].cum_runs_1)
    wicket_runs_2 = list(matchID_df_inns_2[matchID_df_inns_2.index.isin(wicket_index_2)].cum_runs_2)
    
    team_1 = matchID_df_inns_1.batting_team[0]
    team_2 = matchID_df_inns_2.batting_team[0]
    season = matchID_df_inns_1.season[0]
    winner = matchID_df.winner[0]

    plt.figure(figsize=(12,6) , dpi =150)
    plt.plot(matchID_df_inns_1['Ball_No'] ,matchID_df_inns_1['cum_runs_1'], color='b' , label = team_1)
    plt.plot(matchID_df_inns_2['Ball_No'] ,matchID_df_inns_2['cum_runs_2'], color='y' , label = team_2)

    plt.scatter(wicket_index_1, wicket_runs_1, s = 150, color='b')
    plt.scatter(wicket_index_2, wicket_runs_2, s = 150, color='y')

    plt.text(16 , 1, "powerplay")
    plt.text(56, 1, "middle overs")
    plt.text(96 , 1, "death overs")
    plt.title('Interactive visualization of ' + team_1 + ' Vs ' + team_2 + ' in ' + str(season) + '(' +winner+')' )
    
    plt.axvline(36, ls = '--' ,color='grey')
    plt.axvline(90, ls = '--' ,color='grey')

    plt.xlabel('Balls')
    plt.ylabel('Runs')

    plt.legend()

    
    plt.show()
    

    
interactive_viz = interactive_viz(main_df , 1254117)
```
